```python
# Build Youtube object
import googleapiclient.discovery

api_service_name = "youtube"
api_version = "v3"

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=api_key)

youtube
```

```python
# Commnets Done
video_uid = '2fXQvy0kFak'

request = youtube.commentThreads().list(
    part="snippet,replies",
    videoId=video_uid
)


response = request.execute()
lis = []
for item in response['items']:
    msg = item['snippet']['topLevelComment']['snippet']['textDisplay']
    comment_auther = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
#     likes_by_commentor = item['snippet']['topLevelComment']['snippet']['likeCount']
#     comment_date = item['snippet']['topLevelComment']['snippet']['publishedAt']
#     reply_count_to_this_thread = item['snippet']['totalReplyCount']
    lis.append((comment_auther, msg))
lis[0]
```

```python
# Comments option 2 with nested replies
video_uid = '2fXQvy0kFak'

request = youtube.commentThreads().list(
    part="snippet,replies",
    videoId=video_uid
)
response = request.execute()
```

```python
items = response['items']

# top_level_message = item['snippet']['topLevelComment']['snippet']['textDisplay']
# comment_auther = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
all_comments = []
for item in items:
    top_msg = item['snippet']['topLevelComment']['snippet']['textDisplay']
    top_auther = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
    top_date = item['snippet']['topLevelComment']['snippet']['publishedAt']
    replies = []
    if item['snippet']['totalReplyCount'] > 0:
        for reply in item['replies']['comments']:
            msg = reply['snippet']['textDisplay']
            author = reply['snippet']['authorDisplayName']
#             date = reply['snippet']['publishedAt']
            replies.append({'comment_auther':author, 'message':msg})
    data = {'top_msg': top_msg, 'top_author':top_auther, 'replies': replies}
    all_comments.append(data)
```
